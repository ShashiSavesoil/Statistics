```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)  # Ensures reproducibility
x = np.random.randint(1, 100, 50)  # 50 random values between 1 and 100
y = x * 1.5 + np.random.normal(0, 10, 50)  # Linear relation with some noise

# Create scatter plot
plt.scatter(x, y, color='blue', edgecolors='black', alpha=0.7)

# Add labels and title
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.title("Scatter Plot: Relationship Between X and Y")

# Add grid
plt.grid(True, linestyle="--", alpha=0.6)

# Show plot
plt.show()
```
